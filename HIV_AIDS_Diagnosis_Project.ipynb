```python
import pandas as pd
from matplotlib import pyplot as plt
import matplotlib.ticker as mtick
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('/content/HIV_AIDS_DIAGNOSES_and_DEMOGRAPHIC.csv')
df.describe()
```

```python
df.head() # insight on the start of the df
```

```python
df.tail() # insight on the start of the df
```

This dataset from Prof. Rachels Dataset Collection has 11 coloumns and 8900 rows that contain data from 2010 to 2021 on the races, sexes, and neighborhoods diagnosed with HIV and AIDs within the two decade period as well as portion of diagonoses are aids.

```python
plt.figure(figsize=(8,6))
sns.lineplot(data=df, x='YEAR', y='HIV DIAGNOSES PER 100,000 POPULATION', estimator='sum', errorbar=None, color='lightblue')
plt.title('AIDS Diagnoses per 100,000 Population per Year', fontsize=14)
plt.xlabel('Year')
plt.ylabel('HIV diagnosis per 100,00 population')
ax = plt.gca()
ax.yaxis.set_major_locator(mtick.MaxNLocator(10))   # limits number of ticks
ax.yaxis.set_major_formatter(mtick.StrMethodFormatter('{x:,.0f}'))
plt.grid(True, alpha=0.3)
plt.show()
```

From the years 2010 to 2020 HIV diagnosis per 100,000 population significantly increased since after quarantine. There was decrease in 2013 - 2017 with a sharp downward trend during covid-19 peak transmission in 2019-2020. HIV diagnosis per 100,000 doubling post-covid.

```python
plt.figure(figsize=(10,6))
sns.lineplot(data=df, x='YEAR', y='TOTAL NUMBER OF HIV DIAGNOSES', estimator='sum')
sns.lineplot(data=df, x='YEAR', y='TOTAL NUMBER OF AIDS DIAGNOSES', estimator='sum')
plt.title('Total HIV and AIDS Diagnoses per Year', fontsize=14)
plt.xlabel('Year')
plt.ylabel('Total Diagnoses')
ax.yaxis.set_major_locator(mtick.MaxNLocator(10))   # limits number of ticks
ax.yaxis.set_major_formatter(mtick.StrMethodFormatter('{x:,.0f}'))
plt.grid(True, alpha=0.3)
plt.show()
```

This line chart illustrates changes in the total number of HIV diagnoses over a series of years. The x-axis represents time (year), while the y-axis shows the total number of HIV diagnoses. Two lines are displayed to represent different aggregated values in the dataset, allowing for comparison of trends over time. The visualization highlights periods of increase and decrease in diagnoses across the observed time span, including a sharp decline followed by a subsequent increase in the most recent years.

```python
sns.barplot(
    data=df,
    x='SEX',
    y='HIV DIAGNOSES PER 100,000 POPULATION',
    estimator='mean',
    hue='SEX',
    palette='mako',
    legend=False
)
plt.title('HIV Diagnosis Per 100,000 Population By Sex', fontsize=14)
ax = plt.gca()
ax.yaxis.set_major_locator(mtick.MaxNLocator(10))
```

I wanted to see the demographics and boroughs most diagnosed with HIV/ AIDS.

```python
sns.barplot(
    data=df,
    x='RACE/ETHNICITY',
    y='HIV DIAGNOSES PER 100,000 POPULATION',
    estimator='mean',
    hue='SEX',
    palette='rocket'
)
plt.xticks(rotation=45)
ax = plt.gca()
ax.yaxis.set_major_locator(mtick.MaxNLocator(10))
```

This bar chart shows that Black and Hispanic men have the highest rates of HIV diagnoses per 100,000 population, followed by Black and Hispanic women. Across racial and ethnic groups, males consistently exhibit higher diagnosis rates than females. These patterns are observed across the boroughs included in the dataset, highlighting notable disparities by both race/ethnicity and sex.

```python
plt.figure(figsize=(16, 5))
sns.barplot(data=df, x='RACE/ETHNICITY', y='TOTAL NUMBER OF CONCURRENT HIV/AIDS DIAGNOSES',
            estimator='sum', errorbar=None, palette='rocket', hue='RACE/ETHNICITY', legend=False)
plt.grid(True, alpha=0.3)
ax = plt.gca()
ax.yaxis.set_major_locator(mtick.MaxNLocator(10))
plt.title('Total Number of Concurrent HIV/AIDS Diagnoses by Race and Ethnicity', fontsize=14)
plt.show()
```

The bar chart presents the total number of concurrent HIV/AIDS diagnoses categorized by race and ethnicity. It indicates that the total number of concurrent HIV/AIDS diagnoses varies substantially across racial and ethnic groups. Groups such as Black, White, and Latino/Hispanic populations show higher total counts compared to other groups, while Asian/Pacific Islander, Native American, and Multiracial populations show lower totals. These differences might reflect a combination of population size, access to healthcare, testing practices, and underlying social and structural factors.

```python
plt.figure(figsize=(13,6))
sns.barplot(data=df, x='Borough', y='TOTAL NUMBER OF HIV DIAGNOSES', estimator='sum', palette='mako' , hue='RACE/ETHNICITY', legend=True)
plt.title('Number of HIV Diagnosis by Race and Borough', fontsize=14)
plt.xlabel("Borough")
plt.ylabel("Number Of HIV Diagnosis")
ax = plt.gca()
ax.yaxis.set_major_locator(mtick.MaxNLocator(10))
plt.show()
```

This grouped bar chart illustrates the number of HIV diagnoses across New York City boroughs, broken down by race and ethnicity. The x-axis represents the boroughs (All, Bronx, Brooklyn, Manhattan, Queens, and Staten Island), while the y-axis represents the number of HIV diagnoses. Within each borough, multiple bars are shown to represent different racial and ethnic groups, allowing for comparison both within and across boroughs. Both compartive bar charts show black and latino/hispanic groups leading the charts in all boroughs. But this visualization shows Brooklyn and Manhattan leading the chart.
