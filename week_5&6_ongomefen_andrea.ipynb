Task 1: Explain why these columns are important for your analysis.

---

These columns provide detailed information about the passengers, such as age, class, and sex. This data allows us to create distribution graphs (like histograms and KDE plots) to compare how different variables are spread. It also helps visualize relationships and patterns  between variables, giving deeper insight into the factors that may have influenced survival.


Task 2: Histogram and KDE of fare

---



Histogram :

```python

sns.set_palette(sns.light_palette("green", reverse=False))
sns.histplot(data=titanic, x='fare', bins='auto', hue='class', kde=True)

plt.title('Fare Distribution')
plt.xlabel('Fare')
plt.ylabel('Count')
plt.xlim(0,300)
plt.grid(True)

plt.show()
```

KDE:

---



```python
sns.kdeplot(titanic['fare'], fill=True, color='pink')
plt.title('Fare Distribution')
plt.xlabel('Fare')
plt.ylabel('Count')
plt.xlim(0,560)
plt.grid(True)
plt.show()
```

Task 3: Fare and survival distribution comparison

---



```python
sns.kdeplot(data=titanic, x='fare', hue='survived', fill=True, palette='flare')
plt.title('Fare Distribution by Survival')
plt.xlabel('Fare')
plt.ylabel('Density')
plt.grid(True)
plt.show()
```

Task 4: Box plots across multiple groups

---



```python
sns.boxplot(data=titanic, x='pclass', y='fare', hue='sex',palette='crest')
plt.title('Fare Distribution by Passenger Class and Sex')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')
plt.grid(True)
plt.show()
```

As expected, passengers in first class paid the highest fares, followed by those in second and third class, showing a clear link between ticket class and fare amount. Within first class, fares were not only higher but also showed greater variation compared to the other classes. Among first- and second-class passengers, women tended to pay slightly more than men, with the difference being more noticeable in first class. In third class, fare distributions for men and women were similar, though a few female passengers had higher fare outliers.


Task 5:Violin (age by sex).

---



```python
sns.violinplot(data=titanic, x='sex', y='age', palette='flare')
plt.title('Age Distribution by Sex')
plt.xlabel('Sex')
plt.ylabel('Age')
plt.grid(True)
plt.show()
```

The violin plot shows that most passengers were between 20 and 40 years old, where the plot is widest, indicating the highest concentration of individuals. The narrower ends suggest that fewer passengers were in the older age ranges, particularly those around 70 to 80 years old.


Task 6:  Strip (detail on top of violin or boxplot).

---



```python
sns.violinplot(data=titanic, x='sex', y='age', inner=None, color=".8", hue='sex')
sns.stripplot(data=titanic, x='sex', y='age', jitter=0.2, color='lightblue', hue='age')
plt.title('Age Distribution by Sex with Individual Observations')
plt.xlabel('Sex')
plt.ylabel('Age')
plt.grid(True)
plt.show()
```

The violin plot shows the overall density of passenger ages, while the stripplot highlights distinct clusters at specific ages, particularly in the late 20s and early 30s for both sexes. The violin plot provides a smooth estimate of the distribution, whereas the stripplot reveals the actual spread of data, making sparse areas and outliers more visible. For instance, older passengers in their 70s and 80s stand out clearly as individual dots. Additionally, the stripplot exposes gaps where no passengers fall within certain age ranges, details the violin plot might smooth over.
